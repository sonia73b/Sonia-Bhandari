```python
auto=df[['avganncount','avgdeathsperyear','target_deathrate','incidencerate','medincome']]
```

```python
sns.pairplot(auto)
```

```python
#Relationship between the avg income and the target death rate
plt.figure(figsize=(10,5))
sns.scatterplot(data=auto, x='medincome',y='target_deathrate', color='red')
plt.title('Average Income Vs. Target Death Rate')
plt.xlabel('Income')
plt.ylabel('Target Death Rate')
plt.show()
#When we see the below the people who have a average income of about 250000-450000 have higher target death rate of about 150-200.
```

```python
auto1=pd.read_csv('/content/cancer_reg.csv')
```

```python
auto1= auto1.sort_values(by='povertypercent',ascending=False)
top_15_states=auto1.head(15)
plt.figure(figsize=(20,15))
sns.barplot(data=top_15_states, x='povertypercent', y='geography',palette="pastel")
plt.title('Top 15 States in terms of Poverty Percent')
plt.xlabel("Percent of Poverty")
plt.ylabel("State")
plt.show()

#From the first 15 datasets
#the state with the highest poverty percent is Todd County, South Dakota
#the state with the lowest poverty percent is Issaquena County,Mississippi
```

```python
fig, (ax1,ax2,ax3)=plt.subplots(1,3, sharey=True, figsize=(20,10))
ax1.scatter(auto1['pctwhite'],auto1['target_deathrate'],color='red')
ax1.set_title('White Population VS. Target death rate')
ax1.set_xlabel('White Population')
ax1.set_ylabel('Target death rate')


ax2.scatter(auto1['pctblack'],auto1['target_deathrate'],color='blue')
ax2.set_title('Black Population VS. Target death rate')
ax2.set_xlabel('Black Population')
#for the ylabel as it is shared doing it once is enough

ax3.scatter(auto1['pctasian'], auto1['target_deathrate'],color='green')
ax3.set_title('Asian Population VS. Target death rate')
ax3.set_xlabel('Asian Population')

plt.show()

'''we can see that,
White population= 80-100% of the popualtion has about 150-200 target death rate by cancer
Black population= 0-20% the popualtion has about 130-230 target death rate by cancer
Asian population= 0-15% the popualtion has about 120-240 target death rate by cancer'''
#Overall, white popualtion is prone to death being affected by cancer.
```
