```python
from scipy.stats import zscore

#calculating the threshold dynamically
#'data["sulphates"].mean()' = mean of 'sulphates' column
#'data["sulphates"].std()'=standard deviation of 'sulphates' column

z_scores = zscore(data["sulphates"])
z_scores.mean()
threshold = (100 -data["sulphates"].mean() )/data["sulphates"].std()
outliers_mask = abs(z_scores) > threshold

# Removing rows with outliers
data = data[~outliers_mask]
```

```python
data["quality"].value_counts()
```

```python
#'quality' column is mapped in a series
#In this case, the ('3, 4, 5, 6, 7, 8) is replaced by (0,1,2,3,4,5)

data["quality"]=data["quality"].map({3:0, 4:1, 5:2, 6:3, 7:4, 8:5}) #'.map()' = mapping the values in series
```

```python
from sklearn.linear_model import LinearRegression
x_train=df_train[['fixed acidity','volatile acidity','citric acid','residual sugar','chlorides','total sulfur dioxide','density','pH','sulphates','alcohol']]
y_train=df_train['quality']


# for linear regression
lr = LinearRegression()
lr_model = lr.fit(x_train, y_train)


# # #for logistic regression
from sklearn.linear_model import LogisticRegression
lg = LogisticRegression()
lg_model = lg.fit(x_train, y_train)

# # # # #for polynomial regression
from sklearn.preprocessing import PolynomialFeatures
poly_reg = PolynomialFeatures(degree=8)
x_poly = poly_reg.fit_transform(x_train)
# Y_poly = poly_reg.fit_transform(y_train)
pol_reg = LinearRegression()
model = pol_reg.fit(x_poly, y_train)
poly_reg.fit(x_train, y_train)
```

```python
#dumping the model
import pickle as pk
filename= 'model.pickle'
pk.dump(lr_model, open(filename, 'wb'))
```

```python
import joblib  # Used for loading the saved model

def get_wine_data():
    """Collects user input for wine characteristics and returns a DataFrame."""
    try:
        data = {
            'fixed acidity': [float(input("Enter fixed acidity: "))],
            'volatile acidity': [float(input("Enter volatile acidity: "))],
            'citric acid': [float(input("Enter citric acid: "))],
            'residual sugar': [float(input("Enter residual sugar: "))],
            'chlorides': [float(input("Enter chlorides: "))],
            'total sulfur dioxide': [float(input("Enter total sulfur dioxide: "))],
            'density': [float(input("Enter density: "))],
            'pH': [float(input("Enter pH: "))],
            'sulphates': [float(input("Enter sulphates: "))],
            'alcohol': [float(input("Enter alcohol: "))]
        }

        return pd.DataFrame(data)

    except ValueError:
        print("Invalid input. Please enter numerical values for all features.")
        return None

def predict_wine_quality(df):
    "Loads the model, predicts the quality, and categorizes it."
    try:
        # Load your pre-trained model (ensure the path is correct)
        lr_model = joblib.load('/content/model.pickle')

        # Predict the quality
        predicted_quality = lr_model.predict(df)
        rounded_quality = np.round(predicted_quality[0], 2)

        print(f"\nThe Predicted Quality of the wine is: {rounded_quality}")
        return rounded_quality

    except FileNotFoundError:
        print("Model file not found. Please ensure 'model.pickle' exists in the directory.")
        return None
    except Exception as e:
        print(f"An unexpected error occurred: {e}")
        return None

def categorize_quality(quality):
    "Categorizes the wine quality based on the predicted value."
    if quality < 5:
        return "Low"
    elif 5 <= quality <= 7:
        return "Medium"
    else:
        return "High"

if __name__ == "__main__":
    df = get_wine_data()

    if df is not None:
        print("\nInput Data:")
        print(df)

        quality = predict_wine_quality(df)

        if quality is not None:
            quality_category = categorize_quality(quality)
            print(f"The wine is classified as: {quality_category} quality.")
```
